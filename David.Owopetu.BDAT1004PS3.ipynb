```python
# Solution to Step 1 - Import Necessary Libraries

import pandas as pd
import datetime
```

```python
# Solution to Step 2 & 3

file_path = 'wind.txt'

# # Read and store data as read, replace the first three columns as a datetime index
data = pd.read_csv(file_path, sep="\s+", parse_dates={'datetime': [0, 1, 2]})

# # Display the updated dataset
data.head()
```

```python
# Solution to Step 4


# datetime function

def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

# apply the function fix_century on the column and replace the values to the right ones
data['datetime'] = data['datetime'].apply(fix_century)

# data.info()
data.head()
```

```python
# Solution to Step 5

# transform Yr_Mo_Dy it to date type datetime64
data["datetime"] = pd.to_datetime(data["datetime"])

# set 'datetime' as the index
data = data.set_index('datetime')

data.head()
# data.info()
```

```python
# Solution to Step 6

# "Number of non-missing values for each location: "
data.isnull().sum()
```

```python
# Solution to Step 7

#number of columns minus the number of missing values for each location
data.shape[0] - data.isnull().sum()

#or

data.notnull().sum()
```

```python
# Solution to Step 8

data.sum().sum() / data.notna().sum().sum()
```

```python
# Solution to Step 9

data.describe(percentiles=[])
```

```python
# Solution to Step 10

# create the dataframe
day_stats = pd.DataFrame()

day_stats['min'] = data.min(axis = 1) # min
day_stats['max'] = data.max(axis = 1) # max 
day_stats['mean'] = data.mean(axis = 1) # mean
day_stats['std'] = data.std(axis = 1) # standard deviations

day_stats.head()
```

```python
# Solution to Step 11

data.loc[data.index.month == 1].mean()
```

# Question 10
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.
Don't forget to label your axes! Color each dot according to its year.

```python
# Solution to Question 10

file_url = 'arcade-revenue-vs-cs-doctorates.csv'
df = pd.read_csv(file_url)
df.rename(columns = {'Total Arcade Revenue (billions)':'REVENUE','Computer Science Doctorates Awarded (US)':'AWARDS'}, inplace=True)
groups = df.groupby('Year')
for name, group in groups:
    plt.plot(group.REVENUE, group.AWARDS, marker='o', linestyle='', markersize=12, label=name)

plt.legend()
plt.xlabel("REVENUE")
plt.ylabel("AWARDS")

# show the scatter plot
plt.show()
```
